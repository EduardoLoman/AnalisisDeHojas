# Cargar librerías

```python
from scipy import ndimage
import ipywidgets as widgets
import os
import numpy as np
import zipfile
import xmltodict as xml
import pandas as pd
import matplotlib.pyplot as plt
from skimage.morphology import ball
from skimage import measure
```

# Importar tomografías

```python
origen = './Proyectos/'
if os.path.exists(origen) == False:
    os.mkdir(origen)
    
######################
# Cargar tomografías
######################
def save_f(b):
    with outputload:
        print('Cargando Archivos...')
    if upload.value == {}:
        with outputload:
            print('Primero cargue archivos')
        return False
    else:
        with outputload:
            print('Cargando Archivos...')
        for value in upload.value:
            if os.path.exists(origen + value.split('_')[0]) == False:
                os.mkdir(origen + value.split('_')[0])
            if os.path.exists(origen + value.split('_')[0] + '/' + str(int(value.split('_')[1]))) == False:
                os.mkdir(origen + value.split('_')[0] + '/' + str(int(value.split('_')[1])))
            with open(origen + value.split('_')[0] + '/' + str(int(value.split('_')[1])) + '/' + value, 'wb') as f:
                f.write(upload.value[value]['content'])
            with outputload:
                print('Guardado ' + origen + value.split('_')[0] + '/' + str(int(value.split('_')[1])) + '/' + value)
    with outputload:
        print('Archivos cargados')
    upload.value.clear()
    upload._counter = 0
    return True

def refresh(b):
    projects = [f for f in os.listdir(origen) if not f.startswith('.')]
    project_list.options = projects
    files = [int(f) for f in os.listdir(origen + '/' + project_list.value) if not f.startswith('.')]
    if rangep.max < min(files):
        rangep.max = max(files)
        rangep.min = min(files)
    else:
        rangep.min = min(files)
        rangep.max = max(files)
    for i in [0,1]:
        if rangep.max < rangep.value[i]:
            rangep.value[i] = rangep.max
    for i in [1,0]:
        if rangep.min > rangep.value[i]:
            rangep.value[i] = rangep.min
    return True

def extraer(b):
    for direc in range(rangep.value[0], rangep.value[1] +1):
        target = f'{origen}{project_list.value}/{direc}/{project_list.value}_{direc:04}_Mode3D.oct'
        if not os.path.exists(f'{origen}{project_list.value}/{direc}'):
            os.mkdir(f'{origen}{project_list.value}/{direc}')
        if not (os.path.exists(target)):
            with extraer_out:
                print(f'{target} no existe')
            continue
        with extraer_out:
            print('Abriendo archivos...')
        with zipfile.ZipFile(target, 'r') as f:
            header = f.read('Header.xml')
            with open(f'{origen}/{project_list.value}/{direc}/Header.xml', 'wb') as output:
                output.write(header)
            with extraer_out:
                print(f'Guardado {origen}/{project_list.value}/{direc}/Header.xml')
            params = xml.parse(header)['Ocity']['Image']
            with f.open('data/Intensity.data') as intensity:
                vol1 = np.frombuffer(intensity.read(), dtype = np.float32)
            with extraer_out:
                print('Transformando...')
            size = params['SizePixel']
            ImageSize = (int(size['SizeY']), int(size['SizeX']), int(size['SizeZ']))
            vol = vol1.reshape(ImageSize).transpose(0,2,1)
            del vol1
            
            fig, ax = plt.subplots(dpi = 200)
            ax.imshow(vol[int(size['SizeY'])//2])
            ax.axis('off')
            with extraer_out:
                plt.show()
                print('Guardando')
                
            vol = np.where(vol >= 0, vol, 0)
            save_direc = f'{origen}/{project_list.value}/{direc}/Original_{direc}.npz'
            np.savez(save_direc, vol)
            with extraer_out:
                print('Guardado en ' + save_direc)
    return True

upload = widgets.FileUpload(
    accept='',
    multiple=True
)

save = widgets.Button(
    description='Guardar',
    disabled=False,
    button_style='',
    tooltip='Guarda los archivos cargados',
    icon='check'
)
save.on_click(save_f)

outputload = widgets.Output(
    layout={'border': '1px solid white'}
)

upsave = widgets.HBox(
    [upload, save]
)
load = widgets.VBox(
    [upsave, outputload]
)

##############################
# Extraer y guardar imagenes
##############################

projects = [f for f in os.listdir(origen) if not f.startswith('.')]

project_list = widgets.Dropdown(
    options = projects,
    description = 'Proyecto'
)
project_list.observe(refresh)

rangep = widgets.IntRangeSlider(
    value = [0,1],
    step = 1,
    min = 0,
    max =1,
    disabled = False,
    readout=True,
    readout_format='d',
    description = 'Tomografias'
)
rangep.observe(refresh)

extraer_b = widgets.Button(
    description='Extraer',
    disabled=False,
)
extraer_b.on_click(extraer)

extraer_out = widgets.Output()

leer = widgets.VBox(
    [project_list, rangep, extraer_b, extraer_out]
)

tab = widgets.Tab(
    children = [load, leer],
    titles = ['Cargar', 'Convertir']
)
display(tab)
```

# Aplanar superficie

```python
origen = './Proyectos/'
def flat(b):
    flat_button.disabled = True
    output.clear_output()
    dpi = 300
    figsize = np.array([1,3/4])
    for direc in range(mat_start.value, mat_end.value +1):
        path =  f'{origen}{project_list.value}/{direc}/'
        mat_start.value = direc
        with open(f'{path}Header.xml', 'r') as header:
            params = xml.parse(header.read())['Ocity']['Image']
        
        vol = np.load(f'{path}Original_{direc}.npz')['arr_0'].astype('float32')
        shape = vol.shape
        
        print(shape)
        fig, ax = plt.subplots(dpi = dpi)
        ax.imshow(vol[100], aspect = 'auto')
        ax.axis('off')
        with output:
            plt.show()
        
        print(params.keys())
        aspect = float(params['SizeReal']['SizeY'])/float(params['SizeReal']['SizeX'])*shape[2]/shape[0]
        fig_aspect = float(params['SizeReal']['SizeY'])/float(params['SizeReal']['SizeX'])
        
        with output:
            print('Detectando superficie...')
        
        start, end = [10, shape[1]//2]
        
        denoise = ndimage.median_filter(vol[:,start:end], footprint = ball(1)).astype('float32')
        
        q = 99
        
        t = np.quantile(denoise, q/100., axis = 1)
        threshold = denoise >= t[:,None]
        threshold = ndimage.binary_dilation(threshold, iterations = 1)
        labels = measure.label(threshold).astype('uint16')
        max_label, counts = np.unique(labels[labels > 0], return_counts = True)
        df = pd.DataFrame({'labels': max_label, 'counts': counts})
        
        threshold = threshold*0
        for label in df[df['counts'] > shape[1]*shape[2]*2].sort_values('counts')['labels']:
            threshold[labels == label] = 1
            
        threshold = ndimage.binary_dilation(threshold, iterations = 1)
        
        fig, ax = plt.subplots(dpi = dpi)
        ax.imshow(threshold[shape[0]//2], aspect = 'auto')
        ax.axis('off')
        with output:
            plt.show()
        
        surface = np.argmax(threshold, axis = 1).astype('int') + start
        
        del labels
        
        max_surf = surface.max()
        min_surf = surface.min()
        
        vol2 = np.array([[vol[i,(surface[i,j] - 10):surface[i,j] + round(shape[1] - max_surf),j] for j in range(shape[2])] for i in range(shape[0])], dtype =np.float32).transpose(0,2,1)
        shape2 = vol2.shape
        fig, ax = plt.subplots(dpi = dpi)
        ax.imshow(vol2[shape[0]//2], aspect = 'auto')
        ax.axis('off')
        with output:
            plt.show()
        np.savez(f'{origen}{project_list.value}/{direc}/aligned_{direc}.npz', vol2)
        
        del vol, denoise, threshold

        dz = float(params['SizeReal']['SizeZ'])/shape2[1]

        vol2 = ndimage.median_filter(vol2, footprint = ball(1)).astype('float32')

        with output:
            print('Calculando coeficientes de atenuación...')

        vol_sq = vol2**2
        coef = 1/(2*dz)*vol_sq[:,:shape2[1]//2]/((vol_sq[:,::-1]).cumsum(axis = 1)[:,::-1][:,:shape2[1]//2] - vol_sq[:,:shape2[1]//2]).astype('float32')
        
        fig, ax = plt.subplots(dpi = dpi)
        ax.imshow(coef[shape2[0]//2], aspect = 'auto')
        ax.axis('off')
        with output:
            plt.show()
        
        np.savez(f'{origen}{project_list.value}/{direc}/coef_{direc}.npz', coef)
        del vol2, coef, vol_sq
        with output:
            print('Completado')
    flat_button.disabled = False

def refresh_flat(b):
    projects = [f for f in os.listdir(origen) if not f.startswith('.')]
    project_list.options = projects
    files = [int(f) for f in os.listdir(origen + '/' + project_list.value) if not f.startswith('.')]
    if mat_start.max < min(files):
        mat_start.max = max(files)
        mat_end.max = max(files)
        mat_start.min = min(files)
        mat_end.min = min(files)
    else:
        mat_start.min = min(files)
        mat_end.min = min(files)
        mat_start.max = max(files)
        mat_end.max = max(files)
    mat_start.value = max(mat_start.min, mat_start.value)
    mat_start.value = min(mat_start.max, mat_start.value)
    mat_end.value = max(mat_start.min, mat_start.value)
    mat_end.value = min(mat_start.max, mat_start.value)
    flat_button.disabled = False
    mat_start.disabled = False
    mat_end.disabled = False
    return True

projects = [f for f in os.listdir(origen) if not f.startswith('.')]

project_list = widgets.Dropdown(
    options = projects,
    description = 'Proyecto'
)
project_list.observe(refresh_flat)
            
flat_button = widgets.Button(
    disabled = True,
    description = 'Procesar',
    icon = 'play'
)
flat_button.on_click(flat)

mat_start = widgets.BoundedIntText(
    disabled = True,
    value = 0,
    min = 0,
    description = 'Volumen inicial:'
)

mat_end = widgets.BoundedIntText(
    disabled = True,
    value = 0,
    min = 0,
    description = 'Volumen final:'
)

matbox = widgets.HBox(
    [mat_start, mat_end]
)

output = widgets.Output()

box = widgets.VBox(
    [project_list,matbox,flat_button, output]
)
display(box)
```
